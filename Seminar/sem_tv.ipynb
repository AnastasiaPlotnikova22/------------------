## Когда используется формула Пуассона, а когда формула Бернулли?

Формула Бернулли и распределение Пуассона применяются в различных ситуациях в теории вероятностей.

**Формула Бернулли** используется для расчета вероятности того, что событие произойдет ровно \( k \) раз в серии из \( n \) независимых испытаний, где каждое испытание имеет только два возможных исхода (успех или неудача). Это частный случай биномиального распределения, когда вероятность успеха (или неудачи) одинакова в каждом испытании. Формула Бернулли выглядит так:

$$ P(X = k) = C_n^k \cdot p^k \cdot (1-p)^{n-k} $$

где:
- \( P(X = k) \) - вероятность того, что событие произойдет ровно \( k \) раз,
- \( C_n^k \) - число сочетаний из \( n \) по \( k \),
- \( p \) - вероятность наступления события в одном испытании,
- \( n \) - общее количество испытаний,
- \( k \) - количество успешных испытаний, которое нас интересует.

**Распределение Пуассона** применяется для моделирования числа раз, когда событие происходит в фиксированный интервал времени или пространства, при условии, что эти события происходят с постоянной средней скоростью и независимо друг от друга. Это распределение часто используется, когда \( n \) велико, а \( p \) мало, и интересующие нас события редки. Формула распределения Пуассона:

$$ P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!} $$

где:
- \( \lambda \) - среднее число успехов в интервале,
- \( k \) - количество успехов, для которого вычисляется вероятность,
- \( e \) - основание натурального логарифма.

В общем, если испытаний много, а вероятность успеха в каждом испытании мала, и вас интересует вероятность редкого события, то используется распределение Пуассона. Если же испытаний не очень много, или вероятность не очень мала, или вас интересует не только редкое событие, то используется биномиальное распределение, частным случаем которого является формула Бернулли.
