## 1. Из колоды в 52 карты извлекаются случайным образом 4 карты. a) Найти вероятность того, что все карты – крести. б) Найти вероятность, что среди 4-х карт окажется хотя бы один туз.

а)

$$ P(\text{все крести}) = \frac{13}{52} \times \frac{12}{51} \times \frac{11}{50} \times \frac{10}{49} = \frac{11}{4165} \approx 0.00264 $$

Это означает, что вероятность вытянуть все 4 карты крестями из колоды составляет примерно 0.264%.

б)

$$ P(\text{хотя бы один туз}) = 1 - \left( \frac{48}{52} \times \frac{47}{51} \times \frac{46}{50} \times \frac{45}{49} \right) = 1 - \frac{194580}{54145} \approx 0.28126 $$

Таким образом, вероятность того, что среди четырех вытянутых карт окажется хотя бы один туз, составляет примерно 28.126%.

## 2. На входной двери подъезда установлен кодовый замок, содержащий десять кнопок с цифрами от 0 до 9. Код содержит три цифры, которые нужно нажать одновременно. Какова вероятность того, что человек, не знающий код, откроет дверь с первой попытки?

$$ C_{n}^{k} = \frac{n!}{k!(n-k)!} $$

n - общее количество кнопок, а k  - количество кнопок в коде.

$$ C_{10}^{3} = \frac{10!}{3!(10-3)!} = \frac{10 \times 9 \times 8}{3 \times 2 \times 1} = 120 $$

Таким образом, существует 120 различных комбинаций для кода замка.

$$ P(\text{открыть с первой попытки}) = \frac{1}{120} \approx 0.0083 $$

Или примерно 0.83%. Это довольно маленькая вероятность.


## 3. В ящике имеется 15 деталей, из которых 9 окрашены. Рабочий случайным образом извлекает 3 детали. Какова вероятность того, что все извлеченные детали окрашены?

- Общее количество деталей N = 15 
- Количество окрашенных деталей K = 9 
- Количество извлекаемых деталей n = 3 

$$ P(\text{все окрашены}) = \frac{{C_K^n}}{{C_N^n}} $$

C_K^n  - количество сочетаний из  K  по  n , а  C_N^n - количество сочетаний из N  по  n .

$$ P(\text{все окрашены}) = \frac{{C_9^3}}{{C_{15}^3}} = \frac{{\frac{9!}{3!(9-3)!}}}{{\frac{15!}{3!(15-3)!}}} = \frac{{\frac{9 \times 8 \times 7}{3 \times 2 \times 1}}}{{\frac{15 \times 14 \times 13}{3 \times 2 \times 1}}} = \frac{{9 \times 8 \times 7}}{{15 \times 14 \times 13}} $$

$$ P(\text{все окрашены}) = \frac{504}{2730} = \frac{28}{91} \approx 0.3077 $$

Таким образом, вероятность того, что все три извлеченные детали будут окрашены, составляет примерно **30.77%**.


## 4. В лотерее 100 билетов. Из них 2 выигрышных. Какова вероятность того, что 2 приобретенных билета окажутся выигрышными?

- Общее количество билетов N = 100
- Количество выигрышных билетов K = 2
- Количество приобретаемых билетов n = 2


$$ P(\text{оба выигрышные}) = \frac{{C_K^n}}{{C_N^n}} $$

\( C_K^n \) - количество сочетаний из \( K \) по \( n \), а \( C_N^n \) - количество сочетаний из \( N \) по \( n \).

$$ P(\text{оба выигрышные}) = \frac{{C_2^2}}{{C_{100}^2}} = \frac{{\frac{2!}{2!(2-2)!}}}{{\frac{100!}{2!(100-2)!}}} = \frac{1}{{\frac{100 \times 99}{2}}} = \frac{1}{4950} $$

Таким образом, вероятность того, что оба приобретенных билета окажутся выигрышными, составляет **1/4950**, что примерно равно **0.00020202** или **0.0202%**.


# Урок 2. Дискретные распределения вероятностей

## 1. Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

$$ P(X = k) = C_n^k \cdot p^k \cdot (1-p)^{n-k} $$

- \( P(X = k) \) - вероятность того, что событие произойдет ровно \( k \) раз,
- \( C_n^k \) - число сочетаний из \( n \) по \( k \),
- \( p \) - вероятность наступления события в одном испытании,
- \( n \) - общее количество испытаний,
- \( k \) - количество успешных испытаний, которое нас интересует.



 \( p = 0.8 \) (вероятность попадания в мишень),

 \( n = 100 \) (количество выстрелов),
 
 \( k = 85 \) (количество попаданий, которое нас интересует).

$$ P(X = 85) = C_{100}^{85} \cdot 0.8^{85} \cdot (1-0.8)^{100-85} $$

$$ P(X = 85) = C_{100}^{85} \cdot 0.8^{85} \cdot 0.2^{15} $$

Число сочетаний \( C_{100}^{85} \) можно вычислить как:

$$ C_{100}^{85} = \frac{100!}{85!(100-85)!} $$

Теперь вычислим вероятность:

$$ P(X = 85) = \frac{100!}{85! \cdot 15!} \cdot 0.8^{85} \cdot 0.2^{15} $$

Вероятность того, что стрелок попадет в цель ровно 85 раз из 100, составляет примерно **0.048** или **4.8%**.

## 2. Вероятность того, что лампочка перегорит в течение первого дня эксплуатации, равна 0.0004. В жилом комплексе после ремонта в один день включили 5000 новых лампочек. Какова вероятность, что ни одна из них не перегорит в первый день? Какова вероятность, что перегорят ровно две?

- \( \lambda \) (среднее количество событий) = вероятность перегорания одной лампочки \( \times \) общее количество лампочек = \( 0.0004 \times 5000 = 2 \)


$$ P(X = k) = \frac{{e^{-\lambda} \lambda^k}}{{k!}} $$

\( P(X = k) \) - вероятность того, что событие произойдет \( k \) раз.

**Вероятность, что ни одна лампочка не перегорит** (k = 0):

$$ P(X = 0) = \frac{{e^{-2} \cdot 2^0}}{{0!}} = e^{-2} \approx 0.1353 $$

Итак, вероятность, что ни одна лампочка не перегорит в первый день, составляет примерно **13.53%**.

**Вероятность, что перегорят ровно две лампочки** (k = 2):

$$ P(X = 2) = \frac{{e^{-2} \cdot 2^2}}{{2!}} = \frac{{e^{-2} \cdot 4}}{{2}} = 2e^{-2} \approx 0.2707 $$

Следовательно, вероятность, что перегорят ровно две лампочки, составляет примерно **27.07%**.


## 3. Монету подбросили 144 раза. Какова вероятность, что орел выпадет ровно 70 раз?

$$ P(X = k) = C_n^k \cdot p^k \cdot (1-p)^{n-k} $$

где:
- \( P(X = k) \) - вероятность того, что событие произойдет ровно \( k \) раз,
- \( C_n^k \) - число сочетаний из \( n \) по \( k \),
- \( p \) - вероятность наступления события в одном испытании,
- \( n \) - общее количество испытаний,
- \( k \) - количество успешных испытаний, которое нас интересует.

 \( p = 0.5 \) (вероятность выпадения орла),

 \( n = 144 \) (количество подбрасываний),

 \( k = 70 \) (количество раз, когда должен выпасть орел).

Число сочетаний \( C_{144}^{70} \) можно вычислить как:

$$ C_{144}^{70} = \frac{144!}{70!(144-70)!} $$

Теперь вычислим вероятность:

$$ P(X = 70) = \frac{144!}{70! \cdot 74!} \cdot 0.5^{144} $$

Вероятность выпадения орла ровно 70 раз из 144 подбрасываний составляет примерно **0.062** или **6.2%**.
