# Урок 10. Дисперсионный анализ

## 1.
Провести дисперсионный анализ для определения того, есть ли различия среднего роста среди взрослых футболистов, хоккеистов и штангистов. Даны значения роста в трех группах случайно выбранных спортсменов: Футболисты: 173, 175, 180, 178, 177, 185, 183, 182. Хоккеисты: 177, 179, 180, 188, 177, 172, 171, 184, 180. Штангисты: 172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170.

**Шаг 1: Сбор данных**

У нас есть три группы спортсменов с данными о росте:
- Футболисты: $ [173, 175, 180, 178, 177, 185, 183, 182] $
- Хоккеисты: $ [177, 179, 180, 188, 177, 172, 171, 184, 180] $
- Штангисты: $ [172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170] $

**Шаг 2: Тест Бартлетта**

Тест Бартлетта проверяет нулевую гипотезу о том, что все групповые дисперсии равны против альтернативной гипотезы, что хотя бы одна из дисперсий отличается.


**Шаг 3: Выполнение теста в Python**

```python
from scipy.stats import bartlett

# Значения роста спортсменов
footballers = [173, 175, 180, 178, 177, 185, 183, 182]
hockey_players = [177, 179, 180, 188, 177, 172, 171, 184, 180]
weightlifters = [172, 173, 169, 177, 166, 180, 178, 177, 172, 166, 170]

# Проведение теста Бартлетта
stat, p = bartlett(footballers, hockey_players, weightlifters)

# Вывод результатов
print(f"Статистика теста: {stat}, p-значение: {p}")
```

**Шаг 4: Интерпретация результатов**

После выполнения кода мы получим статистическую величину и **p-значение**. Если **p-значение** меньше выбранного уровня значимости (обычно 0.05), мы отвергаем нулевую гипотезу и делаем вывод о наличии статистически значимых различий в дисперсиях. Если **p-значение** больше 0.05, мы не отвергаем нулевую гипотезу и предполагаем, что дисперсии равны.

Эти результаты указывают на то, что p-значение значительно выше уровня значимости 0.05. В контексте теста Бартлетта это означает, что у нас нет достаточных оснований для отклонения нулевой гипотезы о равенстве дисперсий роста футболистов, хоккеистов и штангистов. Следовательно, мы можем предположить, что дисперсии роста во всех трех группах спортсменов статистически не различаются.
